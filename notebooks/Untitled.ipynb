```python
from collections import defaultdict
import pandas as pd
import numpy as np
```

```python
def encode(train, c, t):
    default_ratio = (train[t] > 0).mean()
    general_var = (train[t] > 0).var()
    ct = pd.crosstab(train[c], train[t] > 0)
    ct.columns = [0, 1]
    ct["n"] = ct.sum(axis=1)
    ct["p"] = ct[1] / ct.n
    temp = train.groupby(c)[t].var().fillna(0)
    temp.name = "local_var"
    ct = pd.concat([ct, temp], axis=1)
    ct["pond"] = (ct.n * ct.local_var) / (general_var + ct.n * ct.local_var)
    ct["final"] = ct.pond * ct.p + (1 - ct.pond) * default_ratio
    return defaultdict(lambda: default_ratio, ct.final.to_dict())
```

```python
folds = [["2014-12-31", ("2015-1-01", "2015-3-31")],
         ["2015-3-31", ("2015-4-01", "2015-6-30")],
         ["2015-6-30", ("2015-7-01", "2015-9-30")],
         ["2015-9-30", ("2015-10-01", "2015-12-31")]]
```

```python
train = pd.read_csv("data/01_raw/train.txt", sep=";", na_values="NA",
                    keep_default_na=False, parse_dates=[1])
test = pd.read_csv("data/01_raw/test.txt", sep=";", na_values="NA",
                    keep_default_na=False, parse_dates=[1])
test["returnQuantity"] = -1 
data = pd.DataFrame().append(train)\
                     .append(test)
```

```python
codify_cols = ["articleID"]
```

```python
for i, fold in enumerate(folds):
    for c in codify_cols:
        ct = encode(data[data.orderDate <= fold[0]][[c, "returnQuantity"]], c, "returnQuantity")
        data[c + "_encoded"] =  data[c].apply(lambda x: ct[x]).astype(np.float32)
    
```
